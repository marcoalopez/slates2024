```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
dataset = pd.read_csv('Cijs_decomposed.csv', sep=';')
dataset
```

```python
# estimate the Pearson correlation between calculated elastic constants in transversely isotropic slates
columns = ['C11', 'C33', 'C44', 'C66', 'C13']
corr = dataset[columns].corr()
corr
```

```python
sns.set(font_scale=1.4)
sns.heatmap(corr, cmap='BrBG', vmin=-1, vmax=1, annot=True, linewidths=.6)
```

```python
sns.set(font_scale=1.5)
sns.set_style("ticks")
sns.despine()

g = sns.pairplot(dataset[columns], diag_kind='kde', corner=True, plot_kws = {"s":65})
g.map_lower(sns.kdeplot, levels=4, color="grey")
```
