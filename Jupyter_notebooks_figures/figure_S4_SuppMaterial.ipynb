```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

# Set Nature-like plots
# Figure size
mpl.rcParams['figure.figsize'] = 6.2, 5.0
# Set x axis
mpl.rcParams['xtick.direction'] = 'in'
mpl.rcParams['xtick.major.size'] = 6
mpl.rcParams['xtick.major.width'] = 1
mpl.rcParams['xtick.minor.size'] = 3
mpl.rcParams['xtick.minor.width'] = 1
mpl.rcParams['xtick.minor.visible'] = True
mpl.rcParams['xtick.top'] = True
# Set y axis
mpl.rcParams['ytick.direction'] = 'in'
mpl.rcParams['ytick.major.size'] = 6
mpl.rcParams['ytick.major.width'] = 1
mpl.rcParams['ytick.minor.size'] = 3
mpl.rcParams['ytick.minor.width'] = 1
mpl.rcParams['ytick.minor.visible'] = True
mpl.rcParams['ytick.right'] = True
# Font sizes
mpl.rcParams['axes.labelsize'] = 19
mpl.rcParams['xtick.labelsize'] = 17
mpl.rcParams['ytick.labelsize'] = 17
mpl.rcParams['legend.fontsize'] = 17
mpl.rcParams['font.size'] = 17
# Font Family
mpl.rcParams['font.family'] = 'Open Sans'
# Set line widths
mpl.rcParams['axes.linewidth'] = 1
mpl.rcParams['grid.linewidth'] = 1
mpl.rcParams['lines.linewidth'] = 2.0
mpl.rcParams['lines.markersize'] = 12
# Remove legend frame
mpl.rcParams['legend.frameon'] = False
```

```python
dataset = pd.read_csv('experimental_vs_calc.csv', sep=';')
dataset
```

**Z**: wave propagation normal to slaty cleavage (foliation)  
**Y**: wave propagation parallel to intersection lineation  
**X**: wave propagation normal to intersection lineation within the foliation  
**exp**: experimental Vp and Vs1 measures on dry samples at room conditions  
**calc**: calculated Vp and Vs1 VRH (Hill) wave velocities using the protocol described in Lopez-Sanchez et al. (2024)

```python
# calc ref lines
pct10 = np.linspace(2, 8) + (np.linspace(2, 8) * 0.1)  # 10% faster
pct30 = np.linspace(2, 8) + (np.linspace(2, 8) * 0.3)  # 30% faster
pct50 = np.linspace(2, 8) + (np.linspace(2, 8) * 0.5)  # 50% faster
pct70 = np.linspace(2, 8) + (np.linspace(2, 8) * 0.7)  # 70% faster
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(figsize=(5*2.5, 5), ncols=3, sharey=True, sharex=True, constrained_layout=True)

ax1.plot(np.linspace(2, 8), pct10, linestyle='dotted', color='grey')  # 10% faster
ax1.plot(np.linspace(2, 8), pct30, linestyle='dotted', color='grey')  # 30% faster
ax1.plot(np.linspace(2, 8), pct50, linestyle='dotted', color='grey')  # 50% faster
ax1.plot(np.linspace(2, 8), pct70, linestyle='dotted', color='grey')  # 70% faster
ax1.text(5.6, 6.5, '10% faster', rotation=49, fontsize=13, color='grey')
ax1.text(4.7, 6.5, '30% faster', rotation=53, fontsize=13, color='grey')
ax1.text(4.05, 6.5, '50% faster', rotation=56, fontsize=13, color='grey')
ax1.text(3.5, 6.5, '70% faster', rotation=60, fontsize=13, color='grey')
ax1.plot(dataset['exp_Vp_Z'], dataset['calc_Vp_Z'], 'o', label='Vp')
ax1.plot(dataset['exp_Vs_Z'], dataset['calc_Vs_Z'], 'o', label='Vs1')
ax1.plot([2.0, 8.0], [2.0, 8.0], '-', color='grey')  # reference line
ax1.set_ylabel('predicted (VRH) velocities (km/s)')
ax1.set_title('Axial (Z) direction')
ax1.set_aspect('equal')
ax1.set_xticks([2,3,4,5,6,7,8])
ax1.set_ylim(bottom=2, top=8)
ax1.set_xlim(left=2, right=8)
ax1.grid(which='major', alpha=0.5)
ax1.legend(loc='lower right', fontsize=21)

ax2.plot(np.linspace(2, 8), pct10, linestyle='dotted', color='grey')  # 10% faster
ax2.plot(np.linspace(2, 8), pct30, linestyle='dotted', color='grey')  # 30% faster
ax2.text(4.7, 5.5, '10% faster', rotation=49, fontsize=13, color='grey')
ax2.text(3.9, 5.5, '30% faster', rotation=53, fontsize=13, color='grey')
ax2.plot(dataset['exp_Vp_Y'], dataset['calc_Vp_Y'], 'o', label='Vp')
ax2.plot(dataset['exp_Vs_Y'], dataset['calc_Vs_Y'], 'o', label='Vs1')
ax2.plot([2.0, 8.0], [2.0, 8.0], '-', color='grey')  # reference line
ax2.set_xlabel('laboratory measured velocities (km/s)', fontsize=21)
ax2.set_title('Transverse (Y) direction')
ax2.set_aspect('equal')
ax2.set_xticks([2,3,4,5,6,7,8])
ax2.grid(which='major', alpha=0.5)

ax3.plot(np.linspace(2, 8), pct10, linestyle='dotted', color='grey')  # 10% faster
ax3.plot(np.linspace(2, 8), pct30, linestyle='dotted', color='grey')  # 30% faster
ax3.text(4.7, 5.5, '10% faster', rotation=49, fontsize=13, color='grey')
ax3.text(3.9, 5.5, '30% faster', rotation=53, fontsize=13, color='grey')
ax3.plot(dataset['exp_Vp_X'], dataset['calc_Vp_X'], 'o', label='Vp')
ax3.plot(dataset['exp_Vs_X'], dataset['calc_Vs_X'], 'o', label='Vs1')
ax3.plot([2.0, 8.0], [2.0, 8.0], '-', color='grey')  # reference line
ax3.set_title('Transverse (X) direction')
ax3.set_aspect('equal')
ax3.set_xticks([2,3,4,5,6,7,8])
ax3.grid(which='major', alpha=0.5)
```

_**Figure above**. Comparison between calculated (VRH) Vp and Vs1 wave velocities and those experimentally measured in dry samples under room conditions as a function of the principal directions. This is only for some of the slate samples studied in Cárdenes et al. (2021) and re-analysed here (for more details on the experimental procedures, see Cárdenes et al. 2021). The calculated VRH (Hill) velocities parallel to the foliation (X and Y directions) are ~10% faster than the measured ones, while the measured normal to the slaty cleavage (Z direction) vary widely, being thee calculated ones between ~10 to 70% faster. Note that extrinsic factors leads to an increase in the overall anisotropy of the rock._
